#### After training a LR model, the residual error has the following distribution. Should the model go into production ?

Difficulty: Easy | Automated | Time : 1 min

<center><img src='https://drive.google.com/uc?id=16CtlJai7R9pwLGGv5eOWVpRHZKTGwu4U' width=800></center>



Difficulty: Easy | Automated

Ans: No. The model shouldn't go into production.


**But, why ?**

Recall one of the main assumptions of LR,
- the residuals should be normally distributed.

However, in our case
- the residuals are left skewed.
- i.e. error is not symmetric


We are making more of +ve error.

#### What does it mean when we have +Ve error ? Over-prediction or under-prediction ?

Error is given as : error = $y - \hat{y}$

- So, if error is positive, it means y > ŷ
- or ŷ < y i.e. prediction value is less than actual value
- So, we are under predicting.

#### The model MAPE is 1%. Can we put this model into production?

Difficulty: Medium | Live Open ended | Time : 5 mins

####  What's MAPE ?

<center><img src='https://drive.google.com/uc?id=1DNS_OJdLYCeBjtNLtQtHoc-dPd4UyG2h' width=800></center>


#### Hint :
- Think about the model being used.

#### Answer

We will not put this into production.

Let's understand why.

Notice that
- we are using a tree based regressor to predict the marks.

We understand that how a tree based regressor makes prediction.
- i.e. at leaf node, it'll take mean of all the values present and use it as a predicition.


Now, if the target value lies between 25 -75
- its mean will also lie between 25 - 75.

To conclude
- it can't predict outside the actual target range.

#### How is that an issue ?

Notice the distribution of the target value.


It is been told that
- for training data,
    - mechanics marks lie between 25 - 75.

However, maximum marks for each subject is 100

So, our model will never be able to predict marks < 25 and marks > 75.

#### How to make it work ?

We can instead use a linear regressor
- which focuses on finding the line of best fit instead of box mean.
