# Which sleep phase (REM or Deep) shows higher variability across users? Please explain through data.

```python
deep=df.groupby('User_ID',as_index=False)['master_deep'].std()
rem=df.groupby('User_ID',as_index=False)['master_rem'].std()

q1=deep.merge(rem,on='User_ID')
q1['diff']=q1.master_deep-q1.master_rem
q1.head()
```

```python
sns.histplot(data=q1,x='master_deep')
plt.title('Deep sleep duration histplot')
plt.show()
```

```python
sns.histplot(data=q1,x='master_rem')
plt.title('Rem sleep duration histplot')
plt.show()
```

```python
import scipy.stats as stats

rem = q1["master_rem"]
deep = q1["master_deep"]

levene_stat, p_value_levene = stats.levene(rem,deep)

print(f"Levene Test Statistic: {levene_stat:.4f}")
print(f"P-Value: {p_value_levene:.4f}")


if p_value_levene < 0.05:print("Reject H: Variances are significantly different (Use Welch's t-test).")
else:print("Fail to reject H: Variances are equal (Use standard t-test).")
```

```python
t_stat, p_value = stats.ttest_ind(rem,deep, equal_var=True)  

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:print("Reject H: Deep sleep duration is significantly more than rem sleep duration")
else:print("Fail to reject H: No significant difference in rem or deep sleep duration")
```

# Deep sleep phase shows higher variability across users than rem sleep phase.

# How does stress levels vary based on early sleepers, late sleepers you can assume cutoff for early and late sleep.

```python
df["Circadian midpoint"] = pd.to_datetime(df["Circadian midpoint"], format="%H:%M:%S", errors='coerce')

cutoff_time = pd.to_datetime("02:00:00", format="%H:%M:%S", errors='coerce')   # 2 AM cutoff -> 10PM - 6AM

df["Sleep Category"] = df["Circadian midpoint"].apply(lambda x: "Early Sleeper" if x < cutoff_time else "Late Sleeper")

avg_stress = df.groupby("Sleep Category", as_index=False)["stress_value"].mean()

print(avg_stress.head())
```

```python
import scipy.stats as stats

early_sleeper_stress = df[df["Sleep Category"] == "Early Sleeper"]["stress_value"]
late_sleeper_stress = df[df["Sleep Category"] == "Late Sleeper"]["stress_value"]

levene_stat, p_value_levene = stats.levene(early_sleeper_stress, late_sleeper_stress)

print(f"Levene Test Statistic: {levene_stat:.4f}")
print(f"P-Value: {p_value_levene:.4f}")


if p_value_levene < 0.05:print("Reject H: Variances are significantly different (Use Welch's t-test).")
else:print("Fail to reject H: Variances are equal (Use standard t-test).")
```

```python
early_sleeper_stress = df[df["Sleep Category"] == "Early Sleeper"]["stress_value"]
late_sleeper_stress = df[df["Sleep Category"] == "Late Sleeper"]["stress_value"]

t_stat, p_value = stats.ttest_ind(early_sleeper_stress, late_sleeper_stress, equal_var=False)  

print(f"T-Statistic: {t_stat:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < 0.05:print("Reject H: Early sleepers have significantly lower stress levels.")
else:print("Fail to reject H: No significant difference in stress levels between early and late sleepers.")
```

# Early sleepers have significantly lower stress levels.

# Which weekday had most average sleep recorded.

# Which user_id sleeps most and which user_id sleeps least.

```python
sleep_summary = df.groupby("User_ID")["master_duration"].mean().reset_index()

most_sleep_user = sleep_summary.loc[sleep_summary["master_duration"].idxmax()]
least_sleep_user = sleep_summary.loc[sleep_summary["master_duration"].idxmin()]

print(f"User who sleeps the most: {most_sleep_user['User_ID']} with {most_sleep_user['master_duration']/3600} hours")
print(f"User who sleeps the least: {least_sleep_user['User_ID']} with {least_sleep_user['master_duration']/3600} hours")
```
