```python
import pandas as pd
import datetime as dt
from datetime import datetime, timedelta


#!pip install openpyxl

df=pd.read_excel('transit.xlsx')
date_format = "%y-%m-%d"
df['Date']=pd.to_datetime(df['Date'],format=date_format)
df['Weekday']=df['Date'].dt.day_name()
df.fillna({'One':'one','Two':'iOPEX','Three':'Three','Add':0,'Minus':0, 'yt':0, 'sleep_12':1, 'wake_8':0, 'screen':0, 'Number':95},inplace=True)
```

# BALANCE

```python
current_balance=18000
current_date='24-11-18'
df_=df[df.Date<=pd.to_datetime(current_date)]
print(f'CURRENT BALANCE =======> {current_balance+df_.Add.sum()-df_.Minus.sum()}')
```

# WEIGHT

```python
current_weight = 95
current_date = '24-11-18'
current_date = pd.to_datetime(current_date, format='%y-%m-%d')
days_difference = (dt.datetime.today().date() - current_date.date()).days
expected_weight = current_weight - 0.5 * days_difference/7
print(f'EXPECTED WEIGHT =======> {expected_weight:.2f}')
```

## TODAY

```python
today_date = dt.datetime.today().date()
today_row = df[df['Date'] == pd.to_datetime(today_date)]
today_row
```

## CURRENT WEEK => FIRST WEEK

```python
today_date = datetime.today().date()
start_of_week = today_date - timedelta(days=today_date.weekday())
end_of_week = start_of_week + timedelta(days=6)
start_of_week = pd.to_datetime(start_of_week)
end_of_week = pd.to_datetime(end_of_week)
current_week_rows = df[(df['Date'] >= start_of_week) & (df['Date'] <= end_of_week)]
current_week_rows
```
